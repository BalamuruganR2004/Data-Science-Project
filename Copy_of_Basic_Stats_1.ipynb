## 1. Descriptive Analytics for Numerical Columns

**Objective:** To compute and analyze basic statistical measures (mean, median, mode, standard deviation) for numerical columns.

```python
# Identify Numerical and Categorical Columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print("NUMERICAL COLUMNS:")
print(numerical_cols)
print("\nCATEGORICAL COLUMNS:")
print(categorical_cols)
```

```python
# Descriptive Statistics for Numerical Columns
print("=" * 80)
print("DESCRIPTIVE STATISTICS - NUMERICAL COLUMNS")
print("=" * 80)
df[numerical_cols].describe().round(2)
```

```python
# Calculate Mean, Median, Mode, and Standard Deviation
stats_summary = pd.DataFrame({
    'Mean': df[numerical_cols].mean(),
    'Median': df[numerical_cols].median(),
    'Mode': df[numerical_cols].mode().iloc[0],
    'Std Dev': df[numerical_cols].std(),
    'Min': df[numerical_cols].min(),
    'Max': df[numerical_cols].max()
}).round(2)

print("=" * 80)
print("STATISTICAL MEASURES SUMMARY")
print("=" * 80)
stats_summary
```

- Range: 1 to 15 units

**2. Avg Price:**

**3. Total Sales Value:**

- 75th percentile at ₹57,400 means 75% of transactions are below this value

**4. Discount Rate (%):**
- Average discount: ~13% across all products
- Range: 5% to 20% - indicates a structured discount policy

**5. Net Sales Value:**
- Follows similar pattern to Total Sales Value

## 2. Data Visualization

**Objective:** To visualize the distribution and relationships of numerical and categorical variables.

```python
# Histograms for Numerical Columns
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    axes[i].hist(df[col], bins=20, color='steelblue', edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

    # Add mean line
    axes[i].axvline(df[col].mean(), color='red', linestyle='--', label=f'Mean: {df[col].mean():.2f}')
    axes[i].legend()

plt.suptitle('Histograms of Numerical Variables', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

**1. Volume:**

- Peak at lower volumes indicates majority are small-quantity purchases

**2. Avg Price:**

**3. Total Sales Value & Net Sales Value:**
- Heavily right-skewed distributions
- Most transactions are low-to-medium value
- Long tail indicates occasional high-value transactions (outliers)

**4. Discount Rate (%):**
- Approximately uniform distribution between 5-20%
- No extreme skewness - discounts are evenly applied

**5. Discount Amount:**
- Right-skewed, following the pattern of sales values
- Higher discount amounts correspond to higher-value transactions

## 4. Standardization of Numerical Variables

**Objective:** To scale numerical variables for uniformity using Z-score normalization.

**Concept:** Standardization transforms data to have:
- Mean = 0
- Standard Deviation = 1

**Formula:**
$$z = \frac{x - \mu}{\sigma}$$

Where:
- x = original value
- μ = mean of the feature
- σ = standard deviation of the feature

```python
# BEFORE Standardization - Statistics
print("=" * 70)
print("BEFORE STANDARDIZATION")
print("=" * 70)
print("\nStatistics of Original Data:")
print(df[numerical_cols].describe().loc[['mean', 'std', 'min', 'max']].round(2))
```

```python
# Apply StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numerical_cols])
df_scaled = pd.DataFrame(scaled_data, columns=numerical_cols)

# AFTER Standardization - Statistics
print("=" * 70)
print("AFTER STANDARDIZATION")
print("=" * 70)
print("\nStatistics of Scaled Data:")
print(df_scaled.describe().loc[['mean', 'std', 'min', 'max']].round(6))
```

```python
# Visual Comparison - Before vs After Standardization
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Before Scaling
df[numerical_cols].boxplot(ax=axes[0])
axes[0].set_title('BEFORE Standardization', fontsize=14, fontweight='bold')
axes[0].set_xticklabels(numerical_cols, rotation=45, ha='right')
axes[0].set_ylabel('Original Scale')

# After Scaling
df_scaled.boxplot(ax=axes[1])
axes[1].set_title('AFTER Standardization', fontsize=14, fontweight='bold')
axes[1].set_xticklabels(numerical_cols, rotation=45, ha='right')
axes[1].set_ylabel('Standardized Scale (Z-score)')

plt.suptitle('Before vs After Standardization Comparison', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Histogram Comparison for one variable (Total Sales Value)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Before
axes[0].hist(df['Total Sales Value'], bins=20, color='coral', edgecolor='black', alpha=0.7)
axes[0].set_title('Total Sales Value - BEFORE', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Original Values')
axes[0].set_ylabel('Frequency')
axes[0].axvline(df['Total Sales Value'].mean(), color='red', linestyle='--', label=f'Mean: {df["Total Sales Value"].mean():.0f}')
axes[0].legend()

# After
axes[1].hist(df_scaled['Total Sales Value'], bins=20, color='teal', edgecolor='black', alpha=0.7)
axes[1].set_title('Total Sales Value - AFTER', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Standardized Values (Z-score)')
axes[1].set_ylabel('Frequency')
axes[1].axvline(df_scaled['Total Sales Value'].mean(), color='red', linestyle='--', label=f'Mean: {df_scaled["Total Sales Value"].mean():.4f}')
axes[1].legend()

plt.tight_layout()
plt.show()
```

### Before vs After Standardization - Analysis

**Observations:**

| Metric | Before Standardization | After Standardization |
|--------|------------------------|----------------------|
| Mean | Varies (5 to 41,626) | ≈ 0 for all features |
| Std Dev | Varies (4 to 52,894) | ≈ 1 for all features |
| Scale | Different for each variable | Uniform across all |

**Visual Comparison:**
- **Before:** Variables have vastly different scales - Volume (1-15) vs Net Sales (₹1,000-₹200,000)
- **After:** All variables are now on the same scale, centered around 0

**Why Standardization is Important:**
1. **Equal Feature Contribution:** Prevents high-magnitude features from dominating ML models
2. **Algorithm Requirements:** Many algorithms (SVM, KNN, Neural Networks) require scaled data
3. **Faster Convergence:** Gradient descent converges faster with standardized features
4. **Distance-Based Methods:** Ensures fair distance calculations in clustering algorithms

## 5. Conversion of Categorical Data into Dummy Variables

**Objective:** To transform categorical variables into numerical format using One-Hot Encoding.

**Why One-Hot Encoding?**
- Machine Learning algorithms require numerical input
- Prevents ordinal relationship assumption in nominal categories
- Creates binary (0/1) columns for each category

```python
# Display categorical columns before encoding
print("Categorical Columns to Encode:")
print(categorical_cols)
print(f"\nTotal unique categories across all columns: {sum(df[col].nunique() for col in categorical_cols)}")
```

```python
# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=False)

print("=" * 70)
print("DATASET AFTER ONE-HOT ENCODING")
print("=" * 70)
print(f"\nOriginal Shape: {df.shape}")
print(f"Encoded Shape: {df_encoded.shape}")
print(f"\nNew columns created: {df_encoded.shape[1] - len(numerical_cols)}")
```

```python
# Display sample of encoded data
print("\nSample of Encoded Dataset (first 5 rows, selected columns):")
df_encoded.head()
```

```python
# Show encoding example for one categorical variable
print("=" * 70)
print("ENCODING EXAMPLE: Business Unit (BU)")
print("=" * 70)
print("\nOriginal 'BU' column values:")
print(df['BU'].value_counts())
print("\nAfter One-Hot Encoding:")
bu_cols = [col for col in df_encoded.columns if col.startswith('BU_')]
print(df_encoded[bu_cols].head(10))
```

### One-Hot Encoding Results

**Transformation Summary:**
- Original dataset: 13 columns
- Encoded dataset: 80+ columns (varies based on unique categories)

**Example - Business Unit Encoding:**

| Original BU | BU_FMCG | BU_Lifestyle | BU_Mobiles |
|-------------|---------|--------------|------------|
| Mobiles | 0 | 0 | 1 |
| FMCG | 1 | 0 | 0 |
| Lifestyle | 0 | 1 | 0 |

**Benefits:**
- Each category is now represented as a separate binary feature
- No ordinal relationship is implied between categories
- Ready for use in ML algorithms
